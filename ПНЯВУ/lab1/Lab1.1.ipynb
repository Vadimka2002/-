```python
from numpy import *
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['figure.figsize'] = (4.0, 4.0)

def activation(X, W):
  return 1 / (1 + exp(-summator(W, X)))



def summator(W, X):
  return dot(W,X.T)
def target_func(X,W, y):
  return 0.5 * mean((activation(X,W) - y) ** 2)
def activ_derivative(X, W):
  return activation(X, W) * (1 - activation(X, W))
def target_derivative(X, y, W):
  return ((activation(X, W) - y) * activ_derivative(X, W)).dot(X) / len(y)

def learning(X, W, rate, y, max_steps, eps):
  for i in range(max_steps):
    nabla = target_derivative(X, y, W)
    old_J = target_func(X, W, y)
    W -= rate * nabla
    new_J = target_func(X, W, y)
    if abs(old_J - new_J) < eps:
      return 1
  return 0

X = array([[0, 0, 0], [1, 1, 1], [1, 0 ,1], [1, 1, 0], [0, 1, 1], [0, 0, 1]])
Y = array([1, 1, 1, 0, 1, 0])
W = array([0.0, 0.0, 0.0])

learning(X, W, 0.3, Y, 200, 0.1)

plt.plot(activation(X, W))
plt.title('Функция активации нейрона');
plt.show()

print(activation(array([0, 0, 0]), W) > 0.49)
print(activation(array([1, 1, 1]), W) > 0.5)
print(activation(array([1, 0, 1]), W) > 0.5)
print(activation(array([1, 1, 0]), W) > 0.51)
print(activation(array([0, 1, 1]), W) > 0.5)
print(activation(array([0, 0, 1]), W) > 0.51)
```
