```python
import torch
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['figure.figsize'] = (13.0, 5.0)

x_train = torch.rand(150)
x_train = x_train * 5.0 -3.0

y_train = torch.cos(x_train) / torch.sin(x_train)

plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('$y = ctg(x)$');

```

```python
noise = torch.randn(y_train.shape) / 10
y_train = y_train + noise
plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('noisy ctg(x)')
plt.xlabel('x_train')
plt.ylabel('y_train');
```

```python
x_train.unsqueeze_(1)
y_train.unsqueeze_(1)
tmp = torch.Tensor([1,2,3])
print(tmp)
print(tmp.unsqueeze(1))
```

```python
x_validation = torch.linspace(-2.5, 2.5, 100)
y_validation = torch.cos(x_validation.data) / torch.sin(x_validation.data)
plt.plot(x_validation.numpy(), y_validation.numpy(), 'o')
plt.title('ctg(x)')
plt.xlabel('x_validation')
plt.ylabel('y_validation');
```

```python
x_validation.unsqueeze_(1)
y_validation.unsqueeze_(1);

```

```python
class SineNet(torch.nn.Module):
  def __init__(self, n_hidden_neurons):
    super(SineNet, self).__init__()
    self.fc1 = torch.nn.Linear(1, n_hidden_neurons)
    self.act1 = torch.nn.Sigmoid()
    self.fc2 = torch.nn.Linear(n_hidden_neurons, 1)

  def forward(self, x):
    x = self.fc1(x)
    x = self.act1(x)
    x = self.fc2(x)
    return x

sine_net = SineNet(50)
```

```python
def predict(net, x, y):
  y_pred = net.forward(x)

  plt.plot(x.numpy(), y.numpy(), 'o', label='Ground truth')
  plt.plot(x.numpy(), y_pred.data.numpy(), 'o', c='r', label='Prediction')
  plt.legend(loc='upper left')
  plt.xlabel('$x$')
  plt.ylabel('$y$')

predict(sine_net, x_validation, y_validation)
```

```python
optimizer = torch.optim.Adam(sine_net.parameters(), lr=0.001)
```

```python
def loss(pred, target):
  squares = (pred - target) ** 2
  return squares.mean()
```

```python
for epoch_index in range(700):
  optimizer.zero_grad()

  y_pred = sine_net.forward(x_train)
  loss_val = loss(y_pred, y_train)

  loss_val.backward()

  optimizer.step()

predict(sine_net, x_validation, y_validation)
```
